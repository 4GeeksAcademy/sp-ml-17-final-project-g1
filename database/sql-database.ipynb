# Database Backup

In this notebook we will convert our full_data csv into a **SQLite Database** for backup and easier data management. Our dataset contains almost 500.000 entries with text and corresponding emotion label

```python
import pandas as pd
import sqlite3
```

Importing the **full_data** csv

```python
csv_file = '../data/processed/full_data.csv'
df = pd.read_csv(csv_file)
df.head()
```

Here we create the database and start a connection to it

```python
db_file = 'dataset.sqlite'
conn = sqlite3.connect(db_file)
```

```python
df.to_sql("dataset", conn, if_exists="replace", index=False)
```

Now I try to consult the **Database** to give me 10 anger entries

```python
anger_texts = pd.read_sql("SELECT text, emotion FROM dataset WHERE emotion = 'anger' LIMIT 10", conn)
anger_texts
```

We now close the connection to save memory in our project

```python
conn.close()
```
