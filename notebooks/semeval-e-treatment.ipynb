## Solving the emotions column problem


>To start solving the problem we change the zeros for NaN and the ones for the emotion shown in each column.

```python
emotion_cols = go_emotions.columns.drop('text')

for col in emotion_cols:
    go_emotions[col] = go_emotions[col].replace({1: col, 0: np.nan})

go_emotions.head()
```

>The second step will be to delete the emotions that we are not going to use for our model and merge some emotions into one that can contain them.

```python
go_emotions.replace(['admiration', 'amusement','excitement', 'love', 'optimism'], 'joy', inplace=True)
go_emotions.replace(['grief', 'sadness'], 'sad', inplace=True)
go_emotions.replace(['embarrassment', 'nervousness'], 'fear', inplace=True)
go_emotions.replace(['annoyance', 'disappointment', 'disgust'], 'anger', inplace=True)
not_useful = ['approval', 'caring', 'confusion', 'curiosity', 'desire', 'disapproval', 
              'gratitude', 'pride', 'realization',  'relief', 'remorse', 'neutral']
go_emotions.drop(columns=not_useful, inplace=True)

go_emotions.head()
```

>After changing the name of the emotions we have to merge the data from the columns in the only five we will use in our model and delete the rest of the columns.

```python
go_emotions['joy'].fillna(go_emotions['admiration'], inplace=True)
go_emotions['joy'].fillna(go_emotions['amusement'], inplace=True)
go_emotions['joy'].fillna(go_emotions['excitement'], inplace=True)
go_emotions['joy'].fillna(go_emotions['love'], inplace=True)
go_emotions['joy'].fillna(go_emotions['optimism'], inplace=True)
go_emotions['sadness'].fillna(go_emotions['grief'], inplace=True)
go_emotions['fear'].fillna(go_emotions['embarrassment'], inplace=True)
go_emotions['fear'].fillna(go_emotions['nervousness'], inplace=True)
go_emotions['anger'].fillna(go_emotions['annoyance'], inplace=True)
go_emotions['anger'].fillna(go_emotions['disappointment'], inplace=True)
go_emotions['anger'].fillna(go_emotions['disgust'], inplace=True)
go_emotions.drop(columns=['admiration', 'amusement', 'annoyance', 'disappointment', 'disgust',
       'embarrassment', 'excitement', 'grief', 'love', 'nervousness', 'optimism'],
       inplace=True)
go_emotions.head()
```

>To make the dataframe cleaner and as we don't need more the NaN, we will change them into blank spaces.

```python
go_emotions.replace(np.nan, '', inplace=True)
go_emotions.head()
```

>And now its time to merge it to just one column

```python
final_emotions = go_emotions.columns.drop('text')

go_emotions["emotions_merged"] = go_emotions[final_emotions].apply(
    lambda fila: [val for val in fila if val != ''][0] if sum(fila != '') == 1 
                 else [val for val in fila if val != ""], 
    axis=1
)

go_emotions.drop(final_emotions, axis=1, inplace=True)
go_emotions.head()
```

>As you can see, there is data without any emotions of the ones we are working on and there is also data with multiple basic emotions which means that it is confusing data, so we are deleting all of it using a function to see if it's a list or just a emotion to finish the treatment of the dataset.

```python
def clean_lists(val):
    if isinstance(val, list):
        return np.nan
    else:
        return val

for col in go_emotions.columns:
    go_emotions[col] = go_emotions[col].apply(clean_lists)

go_emotions.dropna(inplace=True)
go_emotions.head()
```
