# SemEval E Treatment

>Here we will treat and process the dataframe semeval-e wich is a subset from SemEval 2018 to work with it with the others dataframes and to make it fit to our model. We will simplyfy it as we did with the others in data-collection.ipynb, leaving it with only two columns, one with the text and the other with the emotion.

```python
import pandas as pd
import numpy as np
```

## Basic simplification
>We will start importing the dataframe and doing as we did with the other dataframes, erasing columns that are useless for this project and a simple general data treatment. And as we did with the other dataframes we will show the top 5 elements.

```python
semeval_e = pd.read_csv('../data/raw/semeval-e.csv')
```

```python
semeval_e.drop(columns=['ID'], inplace=True)
semeval_e.head()
```

>As we can see, we do not have a column for emotions; instead, we have a column for each emotion, where each one contains zeros or ones to indicate whether the emotion is present or not. The next step is to solve this.

## Solving the emotions column problem


>To start solving the problem we change the zeros for NaN and the ones for the emotion shown in each column.

```python
emotion_cols = semeval_e.columns.drop('Tweet')

for col in emotion_cols:
    semeval_e[col] = semeval_e[col].replace({1: col, 0: np.nan})

semeval_e.head()
```

>The second step will be to delete the emotions that we are not going to use for our model and merge some emotions into one that can contain them.

```python
semeval_e.columns
```

```python
semeval_e.replace(['sadness', 'pessimism'], 'sad', inplace=True)
semeval_e.replace(['love', 'optimism'], 'joy', inplace=True)
semeval_e.replace(['disgust'], 'anger', inplace=True)

semeval_e.head()
```

>After changing the name of the emotions we have to merge the data from the columns in the only five we will use in our model and delete the rest of the columns.

```python
semeval_e['joy'].fillna(semeval_e['love'], inplace=True)
semeval_e['joy'].fillna(semeval_e['optimism'], inplace=True)
semeval_e['sadness'].fillna(semeval_e['pessimism'], inplace=True)
semeval_e['anger'].fillna(semeval_e['disgust'], inplace=True)
semeval_e.drop(['anticipation', 'trust', 'love', 'optimism', 'pessimism', 'disgust'], axis=1, inplace=True)
semeval_e.head()
```

>To make the dataframe cleaner and as we don't need more the NaN, we will change them into blank spaces.

```python
semeval_e.replace(np.nan, '', inplace=True)
semeval_e.head()
```

>And now its time to merge it to just one column

```python
final_emotions = semeval_e.columns.drop('Tweet')

semeval_e["emotions_merged"] = semeval_e[final_emotions].apply(
    lambda fila: [val for val in fila if val != ''][0] if sum(fila != '') == 1 
                 else [val for val in fila if val != ""], 
    axis=1
)

semeval_e.drop(final_emotions, axis=1, inplace=True)
semeval_e.head()
```

>As you can see, there is data without any emotions of the ones we are working on and there is also data with multiple basic emotions which means that it is confusing data, so we are deleting all of it using a function to see if it's a list or just a emotion to finish the treatment of the dataset.

```python
def clean_lists(val):
    if isinstance(val, list):
        return np.nan
    else:
        return val

for col in semeval_e.columns:
    semeval_e[col] = semeval_e[col].apply(clean_lists)

semeval_e.dropna(inplace=True)
semeval_e.head()
```

## Data Save
>We are finally saving the dataset treated to the subfolder preprocessed of data.

```python
semeval_e.to_csv("../data/preprocessed/semeval_e.csv", index=False)
```
