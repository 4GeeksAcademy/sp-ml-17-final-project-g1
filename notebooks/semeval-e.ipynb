```python
import pandas as pd
import numpy as np
```

```python
semeval_e_train = pd.read_csv('../data/raw/English/E-c/2018-E-c-En-train.txt', sep='\t')
semeval_e_test = pd.read_csv('../data/raw/English/E-c/2018-E-c-En-test-gold.txt', sep='\t')
semeval_e_dev = pd.read_csv('../data/raw/English/E-c/2018-E-c-En-dev.txt', sep='\t')
semeval_e = pd.concat([semeval_e_train, semeval_e_test, semeval_e_dev])
```

### verificamos si hay columnas que no nos interesa

```python
semeval_e.head()
```

```python
semeval_e.drop(columns=['ID'], inplace=True)
```

```python
semeval_e.head()
```

```python
df= semeval_e.copy()
```

```python
emotion_cols = semeval_e.columns.drop('Tweet')
emotion_cols
```

#### con esto reemplazamos los 1 por las emociones y los 0 desaparecen

```python

for col in emotion_cols:
    df[col] = df[col].replace({1: col, 0: np.nan})

df
```

```python
df.replace(['sadness', 'pessimism'], 'sad', inplace=True)
df.replace(['love', 'optimism'], 'joy', inplace=True)
df.replace(['disgust'], 'anger', inplace=True)
```

```python
df['joy'].fillna(df['love'], inplace=True)
df['joy'].fillna(df['optimism'], inplace=True)
df['sadness'].fillna(df['pessimism'], inplace=True)
df['anger'].fillna(df['disgust'], inplace=True)
df.fillna("", inplace=True )
```

```python
df.drop(['anticipation', 'trust', 'love', 'optimism', 'pessimism', 'disgust'], axis=1, inplace=True)
```

```python
final_emotions = df.columns.drop('Tweet')
```

```python
df["emociones_merged"] = df[final_emotions].apply(
    lambda fila: [val for val in fila if val != ""][0] if sum(fila != "") == 1 
                 else [val for val in fila if val != ""],
    axis=1
)
df.head()
```

### Aqui eliminamos las columnas sobrantes.

```python
df.drop(columns = final_emotions, inplace=True)
```

```python
df.head()
```

### Ahora eliminamos la información incompleta o ambigua

```python
df.emociones_merged.dropna(inplace=True)
```
