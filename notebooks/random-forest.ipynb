# Random Forest test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
import pickle
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
with open('../data/processed/X-train.pkl', 'rb') as file:
    X_train = pickle.load(file)

with open('../data/processed/X-test.pkl', 'rb') as file:
    X_test = pickle.load(file)

with open('../data/processed/y-train.pkl', 'rb') as file:
    y_train = pickle.load(file)

with open('../data/processed/y-test.pkl', 'rb') as file:
    y_test = pickle.load(file)
```

```python
forest = RandomForestClassifier(random_state = 25, class_weight='balanced')
forest.fit(X_train, y_train)
```

```python
y_pred_forest = forest.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_forest)
```

```python
report_forest = classification_report(y_test, y_pred_forest)
print(report_forest)
```

```python
model_confusion_matrix = confusion_matrix(y_test, y_pred_forest)
ConfusionMatrixDisplay(model_confusion_matrix).plot()
plt.show()
```
