```python
import pandas as pd
import numpy as np
```

# Introduction
In this notebook we will treat the go_emotions dataset from google research to fit in our model. Our objective here is to reduce the dataset from the original size to give us only the information we need, in this case the column ['text'] and ['emotion']

## 

## First step:

We load up the data and drop the columns we don't need

```python
go_emotions_1 = pd.read_csv('https://storage.googleapis.com/gresearch/goemotions/data/full_dataset/goemotions_1.csv')
go_emotions_2 = pd.read_csv('https://storage.googleapis.com/gresearch/goemotions/data/full_dataset/goemotions_2.csv')
go_emotions_3 = pd.read_csv('https://storage.googleapis.com/gresearch/goemotions/data/full_dataset/goemotions_3.csv')
go_emotions = pd.concat([go_emotions_1, go_emotions_2, go_emotions_3])
```

```python
go_emotions.loc[go_emotions['example_very_unclear']==True] = np.nan
go_emotions.dropna(inplace=True)
go_emotions.drop(['id', 'author', 'subreddit', 'link_id', 'parent_id',
                  'created_utc', 'rater_id', 'example_very_unclear'], axis =1, inplace=True)
go_emotions
```

## Second Step:
As u can see before emotions are treated as a numerical form in a ton of columns, our goal is to have all the information only in two columns, [text] containing the message and [emotion] containing the emotion

```python

df = go_emotions.copy()
```

```python

text_col = "text"

emotion_cols = go_emotions.columns.drop('text')
```

```python
for col in emotion_cols:
    df[col] = df[col].replace({1: col, 0: ""})

df
```

In the cell below we are doing so text that only have 1 emotion saves as so and texts that have more than 1 emotion saves as a python list

```python
df["emociones_merged"] = df[emotion_cols].apply(
    lambda fila: [val for val in fila if val != ""][0] if sum(fila != "") == 1 
                 else [val for val in fila if val != ""],
    axis=1
)
```

```python
df.drop(emotion_cols, axis=1, inplace=True)
```

## Final Step
We save the treated dataset in .csv

```python
df.to_csv("goemotions_full_clean.csv", index=False)
```
