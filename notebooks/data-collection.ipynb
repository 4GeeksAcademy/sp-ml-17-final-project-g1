# Data Collection

>Here we will collect all the data for the project. It will come from some pre formed dataframes with different intentions and origins like twitter, reddit, and even the sitcom "Friends", adding to this, we will scrap reddit too to also use it in this project.

```python
import pandas as pd
import numpy as np
```

>We'll use data from studies as Crowdflower, GoEmotions, ISEAR, MELD, SemEval, one from KUSHAGRA3204 of Kaggle, Emotion Dataset from Elvis et al. and others, even our own scrap from reddit. You can find more information about this studies in the README.md of data.

>We will start reading that datasets and making one from their splits from each one. You can also find how we scrapped reddit to obtain the dataset called reddit in reddit-scrap.ipynb.

```python
crowdflower = pd.read_csv('../data/raw/crowdflower.csv')

go_emotions_1 = pd.read_csv('https://storage.googleapis.com/gresearch/goemotions/data/full_dataset/goemotions_1.csv')
go_emotions_2 = pd.read_csv('https://storage.googleapis.com/gresearch/goemotions/data/full_dataset/goemotions_2.csv')
go_emotions_3 = pd.read_csv('https://storage.googleapis.com/gresearch/goemotions/data/full_dataset/goemotions_3.csv')
go_emotions = pd.concat([go_emotions_1, go_emotions_2, go_emotions_3])

isear = pd.read_csv('../data/raw/ISEAR.csv')

meld_train = pd.read_csv('../data/raw/MELD/MELD/train_sent_emo.csv')
meld_test = pd.read_csv('../data/raw/MELD/MELD/test_sent_emo.csv')
meld_dev = pd.read_csv('../data/raw/MELD/MELD/dev_sent_emo.csv')

meld_train_dya = pd.read_csv('../data/raw/MELD/MELD_Dyadic/train_sent_emo_dya.csv')
meld_test_dya = pd.read_csv('../data/raw/MELD/MELD_Dyadic/test_sent_emo_dya.csv')
meld_dev_dya = pd.read_csv('../data/raw/MELD/MELD_Dyadic/dev_sent_emo_dya.csv')

meld_train_emorynlp = pd.read_csv('../data/raw/MELD/emorynlp/emorynlp_train_final.csv')
meld_test_emorynlp = pd.read_csv('../data/raw/MELD/emorynlp/emorynlp_test_final.csv')
meld_dev_emorynlp = pd.read_csv('../data/raw/MELD/emorynlp/emorynlp_dev_final.csv')

meld = pd.concat([meld_train, meld_test, meld_dev])
meld_dya = pd.concat([meld_train_dya, meld_test_dya, meld_dev_dya])
meld_emorynlp = pd.concat([meld_train_emorynlp, meld_test_emorynlp, meld_dev_emorynlp])

semeval_EI_anger_train = pd.read_csv('../data/raw/English/EI-oc/training/EI-oc-En-anger-train.txt', sep='\t')
semeval_EI_fear_train = pd.read_csv('../data/raw/English/EI-oc/training/EI-oc-En-fear-train.txt', sep='\t')
semeval_EI_joy_train = pd.read_csv('../data/raw/English/EI-oc/training/EI-oc-En-joy-train.txt', sep='\t')
semeval_EI_sadness_train = pd.read_csv('../data/raw/English/EI-oc/training/EI-oc-En-sadness-train.txt', sep='\t')

semeval_EI_anger_test = pd.read_csv('../data/raw/English/EI-oc/test-gold/2018-EI-oc-En-anger-test-gold.txt', sep='\t')
semeval_EI_fear_test= pd.read_csv('../data/raw/English/EI-oc/test-gold/2018-EI-oc-En-fear-test-gold.txt', sep='\t')
semeval_EI_joy_test = pd.read_csv('../data/raw/English/EI-oc/test-gold/2018-EI-oc-En-joy-test-gold.txt', sep='\t')
semeval_EI_sadness_test = pd.read_csv('../data/raw/English/EI-oc/test-gold/2018-EI-oc-En-sadness-test-gold.txt', sep='\t')

semeval_EI_anger_dev = pd.read_csv('../data/raw/English/EI-oc/development/2018-EI-oc-En-anger-dev.txt', sep='\t')
semeval_EI_fear_dev = pd.read_csv('../data/raw/English/EI-oc/development/2018-EI-oc-En-fear-dev.txt', sep='\t')
semeval_EI_joy_dev = pd.read_csv('../data/raw/English/EI-oc/development/2018-EI-oc-En-joy-dev.txt', sep='\t')
semeval_EI_sadness_dev = pd.read_csv('../data/raw/English/EI-oc/development/2018-EI-oc-En-sadness-dev.txt', sep='\t')

semeval_ei = pd.concat([semeval_EI_anger_dev, semeval_EI_anger_test, semeval_EI_anger_train,
                        semeval_EI_fear_dev, semeval_EI_fear_train, semeval_EI_fear_test,
                        semeval_EI_joy_test, semeval_EI_joy_dev, semeval_EI_joy_train,
                        semeval_EI_sadness_dev, semeval_EI_sadness_test, semeval_EI_sadness_train])

semeval_e_train = pd.read_csv('../data/raw/English/E-c/2018-E-c-En-train.txt', sep='\t')
semeval_e_test = pd.read_csv('../data/raw/English/E-c/2018-E-c-En-test-gold.txt', sep='\t')
semeval_e_dev = pd.read_csv('../data/raw/English/E-c/2018-E-c-En-dev.txt', sep='\t')
semeval_e = pd.concat([semeval_e_train, semeval_e_test, semeval_e_dev])

kushagra = pd.read_csv('../data/raw/combined_emotion.csv')

emotion_dataset_train = pd.read_csv('../data/raw/emotion_training.csv')
emotion_dataset_test = pd.read_csv('../data/raw/emotion_test.csv')
emotion_dataset_validation = pd.read_csv('../data/raw/emotion_validation.csv')
emotion_dataset = pd.concat([emotion_dataset_train, emotion_dataset_test, emotion_dataset_validation])

reddit = pd.read_csv('../data/raw/redditposts.csv')
```

## Basic Treatment

>Before starting with all the data treatment we will do some basic treatment that consists in erease unuseful columns and leave all the dataframes but semeval_e and go_emotions (we will work with them in their own data treatment notebooks due to their complexity) with only two columns that will be the text and the emotion (before processing and normalizing the emotions). We also will be showing the top 5 elements of each simplyfied DataFrame excluded the two mentioned before.

```python
crowdflower.head()
```

>Crowdflower comes simplyfied

```python
isear.drop(['ID'], axis=1, inplace=True)
isear = isear[isear.columns[::-1]]
isear.head()
```

>ISEAR has an useless column named ID and it has turned the text and emotion, so we erased ID and turned the DataFrame

```python
meld.drop(['Sr No.', 'Speaker', 'Sentiment', 'Dialogue_ID', 'Utterance_ID', 'Season', 
           'Episode', 'StartTime', 'EndTime'], axis=1, inplace=True)
meld.head()
```

```python
meld_dya.drop(['Speaker', 'Sentiment', 'Dialogue_ID', 'Utterance_ID', 'Old_Dialogue_ID', 
               'Old_Utterance_ID', 'Season', 'Episode', 'StartTime', 'EndTime'], axis=1, inplace=True)
meld_dya.head()
```

```python
meld_emorynlp.drop(['Speaker', 'Scene_ID', 'Utterance_ID', 'Season', 
                    'Episode', 'Start_Time', 'End_Time'], axis=1, inplace=True)
meld_emorynlp.head()
```

>The three versions of MELD comes with a lot of useless colums to be erased.

```python
for emotion in semeval_ei['Affect Dimension'].unique():
    semeval_ei.loc[semeval_ei['Intensity Class']==f'0: no {emotion} can be inferred'] = np.nan
    semeval_ei.loc[semeval_ei['Intensity Class']==f'1: low amount of {emotion} can be inferred'] = np.nan

semeval_ei.dropna(inplace=True)
semeval_ei.drop(['ID', 'Intensity Class'], axis=1, inplace=True)
semeval_ei
```

>SemEval EI has 4 intensities of emotion so we erased the ones that meant that there was no emotion infered or a low amount, after that we erased the useless columns.

```python
kushagra.head()
```

>Kushagra comes simplyfied

```python
emotion_dataset['label'].replace([0, 1, 2, 3, 4, 5],
                                 ["sadness", "joy", "love", "anger", "fear", "surprise"],
                                 inplace= True)
emotion_dataset.head()
```

>Emotion Dataset from Elvis comes with numbers instead of emotions and what each number meant so we changed the numbers for its own emotion
